**Prediction:** Logistic Regression vs Random Forest Classifier. Which will perform better?

I think random forest will do better because the data has a lot of features, some of which might be a lot of noise.

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
pd.set_option('max_columns', None)
train_df.head()
```

```python
test_df.head()
```

```python
# Get list of category columns
filteredColumns = test_df.dtypes[test_df.dtypes == np.object]
listOfColumnNames = list(filteredColumns.index)
print(listOfColumnNames)
```

```python
for i in listOfColumnNames:
    print(f'Train {i}: {train_df[i].unique()}')
    print(f'Test {i}: {test_df[i].unique()}')
```

```python
#convert target column to numeric

test_df["loan_status"] = test_df["loan_status"].map({'high_risk': 1, 'low_risk': 0})
train_df["loan_status"] = train_df["loan_status"].map({'high_risk': 1, 'low_risk': 0})

train_df['loan_status'].unique()
```

```python
test_df['loan_status'].unique()
```

```python
# Convert categorical data to numeric and separate target feature for training data

train_df2 = pd.get_dummies(train_df)

train_df2
```

```python
# Convert categorical data to numeric and separate target feature for testing data

test_df2 = pd.get_dummies(test_df)

test_df2.head()
```

```python
# add missing dummy variables to testing set

test_df2['debt_settlement_flag_Y'] = 0

test_df2
```

```python
print(len(train_df2.columns))
print(len(test_df2.columns))
```

```python
y_train = train_df2['loan_status']
X_train = train_df2.drop(['loan_status'], axis=1) 
y_test = test_df2['loan_status']
X_test = test_df2.drop(['loan_status'], axis=1 ) 
```

```python
print(X_train)
print(y_train)
print(X_test)
print(y_test)


```

```python
print(f"Train Shape:  {X_train.shape}, {y_train.shape}")
print(f"Test Shape:  {X_test.shape},{ y_test.shape}")
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score (Unscaled Logistic): {model.score(X_train, y_train)}")
print(f"Testing Data Score (Unscaled Logistic): {model.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score

from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(random_state=1, n_estimators=1000)
model2.fit(X_train, y_train)
```

```python
print(f"Training Data Score (Unscaled RandomForest): {model2.score(X_train, y_train)}")
print(f"Testing Data Score (Unscaled RandomForest): {model2.score(X_test, y_test)}")
```
